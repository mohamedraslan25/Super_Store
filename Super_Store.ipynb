# 1-Data Understanding
*Row ID => Unique ID for each row.*<br>

*Order Date => Order Date of the product.*<br>
*Ship Date => Shipping Date of the Product.*<br>
*Ship Mode=> Shipping Mode specified by the Customer.*<br>
*Customer ID => Unique ID to identify each Customer.*<br>
*Customer Name => Name of the Customer.*<br>
*Segment => The segment where the Customer belongs.*<br>
*Country => Country of residence of the Customer.*<br>
*City => City of residence of of the Customer.*<br>
*State => State of residence of the Customer.*<br>
*Postal Code => Postal Code of every Customer.*<br>
*Region => Region where the Customer belong.*<br>
*Product ID => Unique ID of the Product.*<br>
*Category => Category of the product ordered.*<br>
*Sub-Category => Sub-Category of the product ordered.*<br>
*Product Name => Name of the Product.*<br>
*Sales => Sales of the Product.*<br>
*Quantity => Quantity of the Product.*<br>
*Discount => Discount provided.*<br>
*Profit => Profit/Loss incurred.*<br>

**<h4>What are the top selling products in the superstore?<h4>**

```python
product_group1 = df.groupby(["Product Name"]).sum()["Sales"]
product_group1.head()
```

```python
top_selling_values1 = product_group1.sort_values(ascending=False)
top_selling_values1.head()
```

```python
top_selling_values1.head().plot(kind='bar')
plt.title('Top five profit products in superstore')
plt.xlabel('Product Name')
plt.ylabel('Total Price')
plt.show()
```

```python
product_group2 = df.groupby(["Product Name"]).sum()["Profit"]
product_group2.head()
```

```python
top_profit_values2 = product_group2.sort_values(ascending=False)
top_profit_values2.head()
```

```python
top_profit_values2.head().plot(kind='bar')
plt.title('Top five profit products in superstore')
plt.xlabel('Product Name')
plt.ylabel('Total Profit')
plt.show()
```

<h5>"Canon imageCLASS 2200 Advanced Copier" is a top Selling & profit products<h5>

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))
top_selling_values1.head().plot(kind='bar', y='Sales', ax=ax1, color='y')
ax1.set_title('top 5 selling products')

top_profit_values2.head().plot(kind='bar', y='Profit', ax=ax2, color='b')
ax2.set_title('top 5 profit products')

plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.dtypes
```

```python
month_selles = df.groupby(['Order Date'], as_index=False).sum()
month_selles
```

```python
month_selles = month_selles.set_index('Order Date')
```

```python
month_selles = month_selles.resample('M').sum()
```

```python
plt.figure(figsize=(25, 8))
plt.plot(month_selles['Sales'])
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.title('Monthly Sales Trend')
plt.show()
```

```python
year_salles = df.groupby(['Order Date'], as_index=False).sum()
year_salles
```

```python
year_salles = year_salles.set_index('Order Date')
```

```python
year_salles = year_salles.resample('Y').sum()
year_salles
```

```python
plt.figure(figsize=(25, 8))
plt.plot(year_salles['Sales'], color='r')
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.title('Yearly Sales Trend')
plt.show()
```
